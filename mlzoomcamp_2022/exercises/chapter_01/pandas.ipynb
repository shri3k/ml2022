```python
import pandas as pd
```

```python
url = "raw.githubusercontent.com"
path = "/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv"
```

```python
from http import client
conn = client.HTTPSConnection(url)
conn.request("GET", path)
resp = conn.getresponse()
df = pd.read_csv(resp)
```

# Data Accessing

```python
df["Make"], df["Vehicle Size"]
```

```python
df[["Make", "Model", "Year"]]
```

```python
df.index
```

```python
# loc actually uses the index in the table instead of using positional index
df.loc[[1]]
```

```python
# Re-structing the index
import numpy as np
df.index = np.random.randint(low=0,high=len(df), size=(len(df))) # randomizes the indices
df.loc[[1]] # Notice how this is different from above just because we re-structured the index 
```

```python
# To follow the positional index use `iloc`
df.iloc[[1]]
```

# Element-wise oprations

```python
for c in df.columns:
    print(c)
    print(df[c].unique())
    print('\n')
```

```python
df[df["Year"] > 2012] # show all the cars that was manufactured after 2012
```

```python
fuel = df[
    (df["city mpg"] > 50) & 
    (df["Engine Fuel Type"] == "electric")
]
fuel
```
